# Kolmogorov-Smirnov Test (K-S test)

The K-S test may be used to test whether two underlying one-dimensional probability distributions differ. In this case, the K-S statistic is:

$$
D_{n,m} = sup_x|F_{1,n}(x) - F_{2, m}(x)|
$$

Where, $F_{1,n} and F_{2,m}$ are the emperical distribution functions of the first and the second sample respectively, and sup is the supremum function.

<img src="img/kstest.png" width=400 height=200>

The null hypotheses is rejected at level $\alpha$ if

$$
D_{n,m} > c(\alpha)\sqrt{\frac{n+m}{nm}}
$$

where n and m are the sizes of first and second sample respectively. and in general,

$$
c(\alpha) = \sqrt{-\frac{1}{2}ln(\frac{\alpha}{2})}
$$

Note: The two sample test checks whether the two data samples come from the same distribution. This doesnot specify what that common distribution is.

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

#generate a gaussian r.v X
x = stats.norm.rvs(size=1000);
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
stats.kstest(x, 'norm')
```

```python
# Y ~ Continous Uniform Distribution(0,1)
y = np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```
